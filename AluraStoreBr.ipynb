```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que você já carregou os DataFrames das lojas (loja1, loja2, loja3, loja4)

# Função para identificar e plotar os produtos mais e menos vendidos (gráfico de stem)
def plot_top_bottom_products_stem(df, loja_nome, top_n=5, bottom_n=5):
    """
    Identifica e plota os top N e bottom N produtos mais vendidos por quantidade usando um gráfico de stem.

    Args:
        df (pd.DataFrame): DataFrame da loja com colunas 'produto' e 'quantidade'.
        loja_nome (str): Nome da loja para o título do gráfico.
        top_n (int): Número de produtos mais vendidos a serem exibidos.
        bottom_n (int): Número de produtos menos vendidos a serem exibidos.
    """
    vendas_por_produto = df.groupby('Produto')['Preço'].sum().sort_values(ascending=False)

    top_produtos = vendas_por_produto.head(top_n)
    bottom_produtos = vendas_por_produto.tail(bottom_n)

    plt.figure(figsize=(14, 6))

    # Gráfico de stem para os produtos mais vendidos
    plt.subplot(1, 2, 1)
    plt.stem(top_produtos.index, top_produtos.values, linefmt='-', markerfmt='o', basefmt=' ')
    plt.title(f'Top {top_n} Produtos Mais Vendidos - {loja_nome}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    # Gráfico de stem para os produtos menos vendidos
    plt.subplot(1, 2, 2)
    plt.stem(bottom_produtos.index, bottom_produtos.values, linefmt='-', markerfmt='o', basefmt=' ')
    plt.title(f'Top {bottom_n} Produtos Menos Vendidos - {loja_nome}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    plt.show()

# Aplicando a função para cada loja
plot_top_bottom_products_stem(loja1, 'Loja 1')
plot_top_bottom_products_stem(loja2, 'Loja 2')
plot_top_bottom_products_stem(loja3, 'Loja 3')
plot_top_bottom_products_stem(loja4, 'Loja 4')
```

# 6. Desempenho Geográfico

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando o total de vendas por loja
total_vendas_loja1 = loja1['Preço'].sum()
total_vendas_loja2 = loja2['Preço'].sum()
total_vendas_loja3 = loja3['Preço'].sum()
total_vendas_loja4 = loja4['Preço'].sum()

def calcular_vendas_por_local_com_total(df, nome_loja, total_vendas):
    vendas_por_local = df.groupby('Local da compra')['Preço'].sum().reset_index()
    vendas_por_local['Loja'] = nome_loja
    vendas_por_local['Total Vendas Loja'] = total_vendas
    vendas_por_local.rename(columns={'Preço': 'Total Vendas Local'}, inplace=True)
    return vendas_por_local

# Calculando as vendas por 'Local da compra' com o total de vendas da loja
vendas_local_loja1 = calcular_vendas_por_local_com_total(loja1, 'Loja 1', total_vendas_loja1)
vendas_local_loja2 = calcular_vendas_por_local_com_total(loja2, 'Loja 2', total_vendas_loja2)
vendas_local_loja3 = calcular_vendas_por_local_com_total(loja3, 'Loja 3', total_vendas_loja3)
vendas_local_loja4 = calcular_vendas_por_local_com_total(loja4, 'Loja 4', total_vendas_loja4)

# Concatenando os resultados de todas as lojas em um único DataFrame
vendas_por_local_todas_lojas = pd.concat([vendas_local_loja1, vendas_local_loja2, vendas_local_loja3, vendas_local_loja4], ignore_index=True)

# Pivotando a tabela para ter o formato desejado
tabela_pivot = vendas_por_local_todas_lojas.pivot_table(
    index='Local da compra',
    columns='Loja',
    values='Total Vendas Local',
    fill_value=0
)

# Criando uma nova linha para o total de vendas das lojas
total_linha = pd.Series({
    'Loja 1': total_vendas_loja1,
    'Loja 2': total_vendas_loja2,
    'Loja 3': total_vendas_loja3,
    'Loja 4': total_vendas_loja4
}, name='Total')

# Adicionando a linha de total ao DataFrame pivotado
tabela_pivot = pd.concat([pd.DataFrame([total_linha]), tabela_pivot])

# Resetando o índice para ter 'Local da compra' como uma coluna novamente
tabela_pivot.reset_index(inplace=True)
tabela_pivot.rename(columns={'index': 'Local da compra'}, inplace=True)

print(tabela_pivot)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Definindo os locais de compra e as lojas
locais_compra = tabela_pivot['Local da compra']
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criando o gráfico de barras agrupadas
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(locais_compra))

for i, loja in enumerate(lojas):
    plt.bar([pos + i * bar_width for pos in index], tabela_pivot[loja], bar_width, label=loja)

# Adicionando rótulos e título
plt.xlabel('Local da Compra')
plt.ylabel('Total Vendas')
plt.title('Total de Vendas por Local da Compra e Loja')
plt.xticks([pos + bar_width * (len(lojas) - 1) / 2 for pos in index], locais_compra)
plt.legend(title='Loja')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# 1. Faturamento Total por Loja
faturamento_total = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

# 2. Produto Mais Vendido (em quantidade) por Loja
def produto_mais_vendido(df):
    vendas_por_produto = df.groupby('Produto')['Preço'].sum()
    return vendas_por_produto.idxmax()

produtos_mais_vendidos = {
    'Loja 1': produto_mais_vendido(loja1),
    'Loja 2': produto_mais_vendido(loja2),
    'Loja 3': produto_mais_vendido(loja3),
    'Loja 4': produto_mais_vendido(loja4)
}

# 3. Média de Avaliação por Loja
media_avaliacao = {
    'Loja 1': loja1['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

# 4. Custo Médio do Frete por Loja
custo_medio_frete = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

# 5. Vendas por Localização por Loja
def vendas_por_local(df):
    return df.groupby('Local da compra')['Preço'].sum().to_dict()

vendas_localizacao = {
    'Loja 1': vendas_por_local(loja1),
    'Loja 2': vendas_por_local(loja2),
    'Loja 3': vendas_por_local(loja3),
    'Loja 4': vendas_por_local(loja4)
}

faturamento_total = {k: f'{v:.2f}' for k, v in faturamento_total.items()}
media_avaliacao = {k: f'{v:.2f}' for k, v in media_avaliacao.items()}
custo_medio_frete = {k: f'{v:.2f}' for k, v in custo_medio_frete.items()}

# Formatando as vendas por localização para 2 casas decimais
vendas_localizacao = {
    loja: {local: f'{valor:.2f}' for local, valor in vendas.items()}
    for loja, vendas in vendas_localizacao.items()
}

print("Faturamento Total:", faturamento_total)
print("Produto Mais Vendido:", produtos_mais_vendidos)
print("Média de Avaliação:", media_avaliacao)
print("Custo Médio do Frete:", custo_medio_frete)
print("Vendas por Localização:", vendas_localizacao)
```
